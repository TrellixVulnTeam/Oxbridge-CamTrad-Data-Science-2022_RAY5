Complex Graph Example (Annotated)

```python
# Import numpy, a Python library that allows for more direct matrix operations, multi-dimensional arrays, and high-level math
import numpy as np

# Import seaborn, a Python library that allows for more aesthetic and convenient graphing
# seaborn is aliased as "sns" (a fictional character reference)
import seaborn as sns

# Import matplotlib.pyplot, an interface to the Matplotlib library, which allows plotting similar to that of MATLAB
# matplotlib.pyplot is aliased as "plt"
import matplotlib.pyplot as plt

# Sets the theme of the graph to the "dark" preset
sns.set_theme(style="dark")

# Simulate data from a bivariate Gaussian
n = 10000
mean = [0, 0]
cov = [(2, .4), (.4, .2)]
rng = np.random.RandomState(0)
x, y = rng.multivariate_normal(mean, cov, n).T

# Draw a combo histogram and scatterplot with density contours
f, ax = plt.subplots(figsize=(6, 6))
sns.scatterplot(x=x, y=y, s=5, color=".15")
sns.histplot(x=x, y=y, bins=50, pthresh=.1, cmap="mako")
sns.kdeplot(x=x, y=y, levels=5, color="w", linewidths=1)
```
